# Imports

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Extracting Data

```python
prefix = ["100", "250", "500", "2000"]

df = {} 

for file in prefix:
    fileName = "optimisation/" + file + ".txt"
    fileObject = open(fileName, 'r')
    std = []    
    for line in fileObject:          
        part = line.split()
        std.append(float(part[0]))
        
    df[file] = std                
    fileObject.close() 


```

# Plotting

```python
name = ['N=100', 'N=250', 'N=500', 'N=2000']
f = range(3,72,3)
fig, ax = plt.subplots()
for key,val in df.items():
    ax.plot(f, val, "o", label = f"N={key}")
    ax.plot(f, val, "b-")

plt.ylabel("3D standard deviation (mm)")
plt.xlabel("f (%)")
plt.legend()

```
