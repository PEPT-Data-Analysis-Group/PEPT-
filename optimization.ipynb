# Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Extracting Data

```python
prefix = ["100", "250", "500", "2000"]

df = {} # dictionary for storing the dataframes
Nfiles = 26
headerLen = 10


for p in prefix:
    fileName = "optimisation/" + p + ".txt"
    fileObject = open(fileName, 'r')
    u_p = []    
    for l, line in enumerate(fileObject):          
        part = line.split()
        u_p.append(float(part[0]))
        
    df[p] = u_p                
    fileObject.close()
    


```

# Plotting

```python
f = range(3,72,3)
fig, ax = plt.subplots()
for val in df.values():
    ax.plot(f, val, "o")
```
