# Imports

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Extracting Data

```python
prefix = ["100", "250", "500", "2000"]

df = {} 

for file in prefix:
    fileName = "optimisation/" + file + ".txt"
    fileObject = open(fileName, 'r')
    std = []    
    for line in fileObject:          
        part = line.split()
        std.append(float(part[0]))
        
    df[file] = std                
    fileObject.close() 


```

# Plotting

```python
f = range(3,72,3)
fig, ax = plt.subplots()
for val in df.values():
    ax.plot(f, val, "o")
    ax.plot(f, val, "-")
    
```
